```python
#These functions modify data of minimum age and maximum age of fossils, and store the data into
#new CSV file with genus and species data.

infile= "cleaned_Blattodea.csv"
with open(infile, "r") as ff:
     Blattodea_recs=ff.readlines()[0:]
```

```python
from collections import defaultdict
species_ranges=defaultdict(list)

# make a dictionary
for line in Blattodea_recs:
    species=line.split('","')[5]
    minage=line.split('","')[10]
    maxage=line.split('","')[11]
    mean_age=(float(minage) + float(maxage)) /2
    
     # add species as keys and ...
    species_ranges[species].append(mean_age)
```

```python
output=open("Blattodea-ranges.csv","w")

#get the oldest and youngest ages of fossils
#for each species in species_ranges

new_ranges=sorted(species_ranges.keys())

for key in new_ranges:
    ages=species_ranges[key]
    minage=min(ages)
    maxage=max(ages)
    genus=key.split(" ")[0]
    species=key
    #print(genus,key,minage,maxage)
    
    outline="{},{},{},{}\n".format(genus,species,minage,maxage)
    
    print(outline)
    output.write(outline)
```

```python
# This code is used to find coordinates of Blattodea PBDB csv file.
# only longitude and latitude data are stored into the new csv file.
# and the data is used to plot the graph on world map of Rstudio.

infile= "cleaned_Blattodeamap.csv"
with open(infile, "r") as ff:
     Blattodea_recs=ff.readlines()[0:]
```

```python
from collections import defaultdict
species_ranges=defaultdict(list)

for line in Blattodea_recs:
    species=line.split('","')[5]
    lon=line.split('","')[13]
    lat=line.split('","')[14]
    

    species_ranges[species].append(lon)
    species_ranges[species].append(lat)
```

```python
output=open("Blattodeamap-ranges.csv","w")
new_ranges=sorted(species_ranges.keys())

for key in new_ranges:
    edited=species_ranges[key]
    genus=key.split(" ")[0]
    species=key
    lon=edited[0]
    lat=edited[1].replace('"','')
    #print(genus,key,lon,lat)
    
    outline="{},{},{},{}\n".format(genus,species,lon,lat)
    
    print(outline)
    output.write(outline)
```

```python
#Function that takes argument 'name' for name of genus user looking for, and then
#print the row and write on another csv file.
# Newly created CSV file is used to plot the point on the map from Rstudio.

def findgenus(name):
    file=open("Blattodeamap-ranges.csv","r")
    records=file.readlines()
    
    output=open("Periplaneta.csv","w")
    
    for line in records:
        recordgenus= line.split(",")
        if recordgenus[0]==name:
            outline="{},{},{},{}\n".format(recordgenus[0],recordgenus[1],recordgenus[2],recordgenus[3])
            print(outline)           #genus, species, longitude, lattitude data stored in Periplaneta.csv
            
            output.write(outline)
            

findgenus("Periplaneta")
```
